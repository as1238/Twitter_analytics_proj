```python
import tweepy
# Authenticate to Twitter
#We are not including here the real tokens used from the Twitter account that pulled the data fro
auth = tweepy.OAuthHandler("consumer_token", 
    "consumer_secret")
auth.set_access_token("key", 
    "Secret")

api = tweepy.API(auth)

try:
    api.verify_credentials()
    print("Authentication OK")
except:
    print("Error during authentication")
```

```python
# Create a list to append tweet data
tweets_list = list()
tweets_temp = pd.DataFrame()
tweets_list2 = pd.DataFrame()
maxTweets = 100

# Using TwitterSearchScraper to scrape data and append tweets to list
# for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:elonmusk').get_items()):
# for n in range(11):
for a in range(30): 
    dd = datetime.date(2019,12,1)
    d = datetime.date(2019,12,1+a)
#     print(d)
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('$ETH since:'+ str(dd) + ' until:'+ str(d)).get_items()):
        if i>maxTweets:
            break
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
tweets_list2 = pd.DataFrame(tweets_list)
#     tweets_list2 = pd.concat([tweets_list2,tweets_temp])
    
# Creating a dataframe from the tweets list above
tweets_to_df = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```
