```python
# Create a list to append tweet data
tweets_list = list() 
tweets_temp = pd.DataFrame() 
tweets_df2020 = pd.DataFrame() 
maxTweets = 100  

## Product launch dates 
## DPI - Sep 2020 
## GMI - Dec 2021 
## DATA - Sep 2021 
## MVI - June 2021 
```

```python
# Using TwitterSearchScraper to scrape data and append tweets to list
# for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:elonmusk').get_items()):
# for n in range(11):

### PULLING FOR OCT-DEC 2020 

for m in range (10, 13):
    for d in range(30):  
        d1 = datetime.date(2020, m, 1) 
        d2 = datetime.date(2020, m, 1+d) 
    #     print(d) 
        for i,tweet in enumerate(sntwitter.TwitterSearchScraper('$DPI since:'+ str(d1) + ' until:'+ str(d2)).get_items()): 
            if i >= maxTweets: 
                break 
            tweets_list.append([tweet.date, tweet.id, tweet.content]) 
```
